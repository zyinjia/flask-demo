```python
df1 = df1.rename(columns={'data':'date','open':'open'})
date0 = df1['date'][0]
```

```python
from datetime import datetime
def deltadays(unidate):
    # input is a unicode format date
    # return an int of the days difference compared to today
    strpdate = datetime.strptime(unidate, '%Y-%m-%d')
    nowdate = datetime.today()
    delta = nowdate - strpdate
    return delta.days

def getdatetime(unidate):
    # input is a unicode format date
    # return a datetime object
    strpdate = datetime.strptime(unidate, '%Y-%m-%d')
    return strpdate
```

```python
df1.head()
```

```python
df1['deltadays'] = [deltadays(i) for i in df1['date']]
df1['datetime'] = [getdatetime(i) for i in df1['date']]
df1.head()
```
